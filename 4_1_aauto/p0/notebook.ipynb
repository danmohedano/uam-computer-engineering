Para comprobar el funcionamiento de la Validación Simple, se crea una estrategia con una proporción de test del 20% y 2 ejecuciones. Se muestra entonces el contenido de ambos índices en las particiones creadas.

```python
import EstrategiaParticionado
simple = EstrategiaParticionado.ValidacionSimple(0.2, 2)
particiones = simple.creaParticiones(dataset, 123)
for p in particiones:
    print('PARTICIÓN {}'.format(particiones.index(p)))
    print('#TEST: {}\n#TRAIN: {}\n'.format(p.indicesTest, p.indicesTrain))
```

Para comprobar el funcionamiento de la Validación Cruzada, se repite lo mismo. En este caso, se le pide generar 10 particiones.

```python
cruzada = EstrategiaParticionado.ValidacionCruzada(10)
particiones = cruzada.creaParticiones(dataset, 123)
for p in particiones:
    print('PARTICIÓN {}'.format(particiones.index(p)))
    print('#TEST: {}\n#TRAIN: {}\n'.format(p.indicesTest, p.indicesTrain))
```
