```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python

def plot_data(filenames, title, legend):
    dfs = [pd.read_csv(filename, delimiter='\t', error_bad_lines=False, skipfooter=1, engine='python') for filename in filenames]
    
    fig = plt.figure(figsize=(10,7))
    gs = fig.add_gridspec(2, hspace=0.05)
    ax1, ax2 = gs.subplots(sharex=True)

    fig.suptitle(title)
    
    [ax1.plot(df.iloc[:, 0].values, df.iloc[:, 1].values) for df in dfs]
    ax1.set_ylabel('ECM')
    ax1.legend(legend)
    
    [ax2.plot(df.iloc[:, 0].values, df.iloc[:, 2].values) for df in dfs]
    ax2.set_ylabel('Accuracy')
    ax2.legend(legend)
    ax2.set_xlabel('Epochs')
     
    
```

```python
files = ['perceptron_alpha_1.dat', 'perceptron_alpha_0.1.dat', 'perceptron_alpha_0.01.dat', 'perceptron_alpha_0.001.dat']
plot_data(files, r"Perceptron con $\theta = 0.2$", [r'$\alpha = 1$', r'$\alpha = 0.1$', r'$\alpha = 0.01$', r'$\alpha = 0.001$'])

files = ['perceptron_umbral_0.1.dat', 'perceptron_umbral_0.2.dat', 'perceptron_umbral_0.3.dat', 'perceptron_umbral_0.4.dat']
plot_data(files, r"Perceptron con $\alpha = 0.01$", [r'$\theta = 0.1$', r'$\theta = 0.2$', r'$\theta = 0.3$', r'$\theta = 0.4$'])

files = ['adaline_alpha_0.1.dat', 'adaline_alpha_0.01.dat', 'adaline_alpha_0.001.dat']
plot_data(files, r"Adaline con $tol = 0.01$", [r'$\alpha = 0.1$', r'$\alpha = 0.01$', r'$\alpha = 0.001$'])

files = ['adaline_tol_0.1.dat', 'adaline_tol_0.01.dat', 'adaline_tol_0.001.dat', 'adaline_tol_0.0001.dat']
plot_data(files, r"Adaline con $\alpha = 0.01$", [r'$tol = 0.1$', r'$tol = 0.01$', r'$tol = 0.001$', r'$tol = 0.0001$'])
```
