```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_des = pd.read_csv('../test/des/histograma.csv', sep='\t').to_numpy()
data_des_sbox = [data_des[i : i + 31] for i in range(0, 250, 31)]
data_aes = pd.read_csv('../test/aes/histograma.csv', sep='\t').to_numpy()
data_aes_sbox = [data_aes[:512], data_aes[512:]]
```

```python
def hist(axis, data, title):
    values = [x[1] for x in data]
    bins = [x[0] for x in data]
    axis.bar(bins, values)
    axis.set_xlabel("$S(A\oplus B) - S(A)\oplus S(B)$")
    axis.set_title(title)
```

```python
fig, axs = plt.subplots(2, 4, figsize=(20,10))
fig.subplots_adjust(hspace=0.25, wspace=0.2)
for i in range(8):
    hist(axs.flatten()[i], data_des_sbox[i], f'S-BOX {i+1}')

fig.savefig('../test/des/histograma_des.png')
```

```python
fig, axs = plt.subplots(2,1, figsize=(15, 15))
aes_titles = ['AES Direct S-BOX', 'AES Inverse S-BOX']
fig.subplots_adjust(hspace=0.2)

for i in range(2):
    hist(axs.flatten()[i], data_aes_sbox[i], aes_titles[i])
fig.savefig('../test/aes/histograma_aes.png')
```
